```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('heart.csv')
print(df.head())
print(df.describe())
print(df.info())
print(df.isnull())
df['age'].plot(kind='hist', bins=15, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
df['sex'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
print(df.size)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('USA_cars_datasets.csv')
print(df.head())
print(df.describe())
print(df.info())
print(df.isnull())
print(df.shape)
print(df.size)
plt.figure(figsize=(8, 4))
sns.histplot(df['price'], bins=25)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
